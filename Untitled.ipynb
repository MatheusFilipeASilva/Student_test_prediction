```python
import platform
import os
import psutil

vm = psutil.virtual_memory()
disk = psutil.disk_usage("/")

print(f"RAM total: {vm.total / 1024**3:.2f} GB")
print(f"RAM disponível: {vm.available / 1024**3:.2f} GB")
print(f"Uso de RAM: {vm.percent}%")

print(f"Disco total: {disk.total / 1024**3:.2f} GB")
print(f"Disco livre: {disk.free / 1024**3:.2f} GB")
```
